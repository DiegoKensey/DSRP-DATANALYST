````python
data = {
          "masa":np.random.randint(45,120,100000), # en kg
          "talla":np.random.randint(150, 210, 100000)/100,#en metros
          "edad": np.random.randint(18, 85, 100000)
        }
df_personas = pd.DataFrame(data)
df_personas
````

Deseo generar una variable categorica en función de la edad que indique si es adulto mayor o no la persona.
* Adulto mayor (>=58) # True
* Joven               # False


* Forma 1: utilizando `apply(), que nos permite definir una función y aplicarla sobre toda la columna. 
Pasos:
  *  Crear la función que recibe como entrada el valor deseado
  *  Utilizar `apply()` para la columna

````python
def es_adultomayor(edad):
  if edad>=58:
    return True
  return False
````

````python
%%time
df_personas["AdultoMayor"] = df_personas["edad"].apply(es_adultomayor)
````

````python
df_personas.head()
````

Tener cuidado con `apply()` pues emula lo que hace un for

````python
def es_adultomayor2(column):
  return column>=58
````

````python
%%time
df_personas["AdultoMayor"] = es_adultomayor2(df_personas["edad"])
````

## **2.7 Groupby**

<img src="https://miro.medium.com/max/1400/1*KT4Zn8JTjcjzjCR2NFxYfg.png" alt ="none" width= "600"  height="350">

Se utiliza para organizar datos identicos en grupos. El grupo se crea dividiendo los datos de acuerdo con las categorias de las columnas seleccionadas



````python
df3 = pd.DataFrame(
    {
      "clave1":list("aabbababb") ,
      "clave2": ["uno", "dos", "tres"]*3,
      "datos1": np.random.randn(9),
      "datos2": np.random.randn(9),
    }
)
df3
````

Tenemos `clave1 `y `clave2` como variables categoricas

````python
grupo = df3["datos1"].groupby(df3["clave1"])
grupo
````

````python
grupo.max()
````

````python
grupo2 = df3["datos1"].groupby([df3["clave1"], df3["clave2"]])
grupo2.mean()
````

````python
grupo2.mean().unstack()
````

````python
df3.groupby("clave1").mean()
````

````python
df3.groupby(["clave1", "clave2"]).mean()
````

````python
for nombre, grupo in df3.groupby("clave1"):
  print(f"nombre:{nombre}")
  print(f"Grupo:{grupo}")
````

````python
for (n1,n2), grupo in df3.groupby(["clave1", "clave2"]):
  print(f"nombre1:{n1}, nombre2:{n2}")
  print(f"Grupo:{grupo}")
  print(type(grupo))
````

````python
partes = dict(list(df3.groupby("clave1")))
partes["a"]
````

````python
partes["b"]
````

se puede seleccionar una parte del grupo

````python
df3.groupby(["clave1", "clave2"]).mean()
````

````python
df3.groupby(["clave1", "clave2"])[["datos1"]].count()
````

## **2.1 Pivot_table**

````python
df_churn = pd.read_csv("/content/drive/MyDrive/Datasets/datasets DSRP/PEADP /PythonAnalisisdedatos/Churn.csv")
df_churn.head()
````

````python
df_churn.info()
````

````python
df_churn.pivot_table(
    "number vmail messages",  # contenido
    index = "state",          # fila
    columns = "international plan", # columna
    aggfunc= "sum"            #operador
)
````

````python
df_churn.pivot_table(
    ["number vmail messages", "total day minutes"],  # contenido
    index = ["state", "churn"],          # fila
    columns = ["international plan", "voice mail plan"], # columna
    aggfunc= "sum" , #operador
    fill_value = 0
)
````

````python
df_churn
````

````python
df_churn.pivot_table(
    ["number vmail messages", "total day minutes"],  # contenido
    index = ["state", "churn"],          # fila
    columns = ["international plan", "voice mail plan"], # columna
    aggfunc= {"total day minutes":"sum",
              "voice mail plan":"max"},        #operador
    fill_value = 0
)
````
